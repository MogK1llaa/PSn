```python
import geopandas as gpd
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import os

filePath='D:/PSn/gadm36_PRT.gpkg'
path1='D:/PSn/'

shapefile = gpd.read_file(filePath)
shapefile['join']=shapefile['NAME_1']+shapefile['NAME_2']
shp=shapefile.dissolve(by='join')
```

```python
exc=pd.read_excel('data.xlsx', index_col='Unnamed: 0')
exc.index=[v.strip() for v in exc.index]

munExcel = exc.drop('Portugal').index.unique().to_list()

munGeo=shp['NAME_2'].to_list()

shp=shp.rename(columns={'NAME_1':'Distrito'})
shp.loc[shp['NAME_2']=='Ponte de Sôr','NAME_2']='Ponte de Sor'
shp.loc[shp['NAME_2']=='Praia da Vitória','NAME_2']='Vila da Praia da Vitória'
shp.loc[((shp['NAME_2']=='Calheta') & (shp['Distrito']=='Azores')),'NAME_2']='Calheta [R.A.A.]'
shp.loc[((shp['NAME_2']=='Calheta') & (shp['Distrito']=='Madeira')),'NAME_2']='Calheta [R.A.M.]'
shp.loc[((shp['NAME_2']=='Lagoa') & (shp['Distrito']=='Azores')),'NAME_2']='Lagoa [R.A.A.]'

shp.index=shp['NAME_2']

madeira=shp[shp['Distrito']=='Madeira'].index.to_list()

acores=shp[shp['Distrito']=='Azores'].index.to_list()
continente = shp[(shp['Distrito']!='Azores') & (shp['Distrito']!='Madeira')].index.to_list()

exc['Distrito']=shp['Distrito']
exc['geometry']=shp['geometry']

#for m in munExcel:
 #   if m not in munGeo:
  #      print('Falta',m)
#print('**')
#for m in munGeo:
#    if m not in munExcel:
#        print('Falta',m)
```

```python
import random as rand
def printMap(df, area, title):
    cmap = plt.cm.rainbow
    
    ax = plt.gca()
    for m in area:
        muni=gpd.GeoDataFrame(df.loc[m].to_frame().T)
        value=rand.random()
        colorCode=mpl.colors.to_hex(cmap(value)[0:3])
        muni.plot(color=colorCode, ax=ax)
    plt.title('Madeira')
    plt.show()
    
printMap(exc[exc['Ano']==2010],acores,'Mapa dos Açores') 
printMap(exc[exc['Ano']==2010],madeira,'Mapa da Madeira') 
printMap(exc[exc['Ano']==2010],continente,'Mapa do Continente') 
```

```python
def colorFader(c1,c2,mix=0):
    c1=np.array(mpl.colors.to_rgb(c1))
    c2=np.array(mpl.colors.to_rgb(c2))
    return mpl.colors.to_hex((1-mix)*c1+mix*c2)
```

```python
def makeMaps(df,col):
    
    cmap = plt.cm.rainbow
    
    ax = plt.gca()
    for m in madeira:
        muni = municipios['geometry'][municipios['NAME_2']==m]
        value = df.loc[m][col]
        colorCode=mpl.colors.to_hex(cmap(value)[0:3])
        muni.plot(color=colorCode, ax=ax)
    plt.title('Madeira'+str(df['Ano'][0]))
    plt.show()
    
    ax = plt.gca()
    for m in acores:
        muni=municipios['geometry'][municipios['NAME_2']==m]
        value=df.loc[m][col]
        colorCode=mpl.colors.to_hex(cmap(value)[0:3])
        muni.plot(color=colorCode, ax=ax)
    plt.title('Açores'+str(df['Ano'][0]))
    plt.show()
    
    ax = plt.gca()
    for m in continente:
        muni=municipios['geometry'][municipios['NAME_2']==m]
        value=df.loc[m][col]
        colorCode=mpl.colors.to_hex(cmap(value)[0:3])
        muni.plot(color=colorCode, ax=ax)
    plt.title('Continente'+str(df['Ano'][0])
    plt.show()
```
