```python
import pandas as pd
import os

path='E:/faculdade/PSn_Municipios/excelfiles/'
df = pd.read_excel(os.path.join(path,'O seu município em números! Abrantes.xls')
                  )
df=df[6:78]
df=df.drop(['Unnamed: 1','Unnamed: 3','Unnamed: 5','Unnamed: 6'], axis=1)

df.columns=['Municipio']+df.loc[7][1:].tolist()
df.loc[6]=['Ano',2010,2019]

df=df.drop([7]).reset_index(drop=True)
df=df.transpose()
df.columns=df.loc['Municipio']
df=df.drop(['Municipio'])
```

```python
f=open('E:/faculdade/PSn_Municipios/municipios.txt','r', encoding='utf8')
listaMun= [mun.strip() for mun in f.readlines()]
for mun in listaMun:
    print(f'O seu município em números! {mun}.xls')
    print('***************')
    df1 = pd.read_excel(os.path.join(path,f'O seu município em números! {mun}.xls'))
    df1=df1[6:78]
    df1=df1.drop(['Unnamed: 2','Unnamed: 4','Unnamed: 5','Unnamed: 6'], axis=1)
    df1.columns=['Municipio']+df1.loc[7][1:].tolist()
    df1.loc[6]=['Ano',2010,2019]
    df1=df1.drop([7])
    df1=df1.transpose()
    df1.columns=df1.loc['Municipio']
    df1=df1.drop(['Municipio'])
    df1.columns=df.columns
    df=df.append(df1) 
    
df.columns=[c.strip() for c in df.columns]    
df= df.drop (['Empresas não financeiras','€, milhares','Superfície em km2'], axis=1)

for c in df.columns:
    if df[c].isnull().all():
        df=df.drop(c,axis=1)
df.to_excel('data.xlsx')
```
